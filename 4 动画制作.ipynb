编写一小段跳舞指令，然后组合跳舞指令，形成更丰富的舞蹈动作。

```python
# 组合舞步
import time
from codelab_adapter_client.message import send_message

send_message('播放音乐')
time.sleep(0.5)
d1 = ['上', '下', '上', '右']
d2 = ['左', '下', '左', '右']
d3 = ['上', '下', '左', '右']
d4 = ['左', '下', '上', '右']
dance = [d1, d2, d3, d4]

for d in dance:
    for move in d:
        send_message(move)
        time.sleep(0.5)
send_message('停止音乐')
```

现在我们来制作一个有节奏感的走路动画，打开项目[【交互计算】走路](https://create.codelab.club/projects/8803/editor)，用左右键控制角色移动，体验有节奏地走路。使用 Python 编程，通过设定前进或后退的次数，形成有节奏的走路动画。

```python
#【交互计算】走路 https://create.codelab.club/projects/8803/editor/
import time
from codelab_adapter_client.message import send_message

steps = [2, -2, 1, -1, 5, -5]
msg = '前进'
for step in steps:
    if step > 0:
        msg = '前进'
    else:
        msg = '后退'
    for i in range(abs(step)): # abs() 可以获得绝对值
        send_message(msg)
        time.sleep(0.2)
```

### 编程百科
#### 条件判断
这里的程序用到了条件判断，数字代表移动的次数，数字为正，表示前进，数字为负，代表后退。Python 中的条件判断和 Scratch 是类似的。



#### 比较运算符
这里的条件判断用到了比较运算符，要特别注意等于和不等于运算符的使用。

| 比较运算符 | 说明 | 
|------|------|
|  >  | 大于 |
|  <  | 小于 |
|  >=  | 大于等于 |
|  <=  | 小于等于 |
|  ==  | 等于 |
|  !=  | 不等于 |

进行条件判断时，如果条件成立（True），则条件判断中的代码执行。不过这里有很多种方式让条件成立。

```python
print(6 > 5)
if 2:
    print(2)
if 'hello':
    print('hello')
```

当条件判断中的表达式的值不是 0、False 或 None 时，条件都会成立。True 和 False 是布尔值。

```python
if 0:
    print(0)
if False:
    print(False)
if None:
    print(None)
```

也可以单独使用 if 条件判断，如果有多个并列的条件判断，则可以使用 if-elif-else 的形式。

```python
a = 6
if a > 3:
    print('a 大于 3')
    
b = 9
if b < 6:
    print('b 小于 6')
elif b > 10:
    print('b 大于 10')
else:
    print('b 大于等于 6，小于等于 10')
```
