### 编程百科
#### 数学运算符
Python 中的数学运算符和我们日常使用的数学运算符有一些区别，使用 Python 进行数学计算时要注意。

#### 数字类型
Python 中的常见的数字类型是**整数**(integer)和**浮点数**(float)，浮点数也就是数学中的小数。不同类型的数字进行运算时，要注意结果的类型。整数和浮点数运算的结果是浮点数。

```python
a = 5 + 3
b = 5.0 + 3
print(a)
print(b)
```

## 进阶-math 库
如果要进行更加复杂的数学计算，那么可以使用 Python 内置的 math 库，这里给出一些常见的函数。如果想了解更多 math 库的知识，可以查看[官方文档](https://docs.python.org/zh-cn/3/library/math.html)。

| 函数 | 描述 |
|-----|-----|
|  math.ceil(x)  | 返回 x 的上限，即大于或者等于 x 的最小整数。 |
|  math.floor(x)  | 返回 x 的向下取整，即小于或等于 x 的最大整数。 |
|  math.pow(x, y)  | 返回 x 的 y 次幂。 |
|  math.sqrt(x)  | 返回 x 的平方根。 |
|  math.sin(x) | 返回 x 弧度的正弦值。 |
|  math.cos(x) | 返回 x 弧度的余弦值。 |
|  math.tan(x) | 返回 x 弧度的正切值。 |

```python
# math 库示例
import math

math.sqrt(9)
```
