```python
# https://create.codelab.club/projects/10621/editor/
from codelab_adapter_client.message import send_message, receive_message

coordinates = []
while True: 
    msg = receive_message()
    if msg:
        print(msg)
        msg = msg.split(',')
        if msg[0] == '':
            coordinates.append(msg[1])
            coordinates.append(msg[2])
        elif msg[0] == '':
            # 
            coordinates.remove(msg[1])
            coordinates.remove(msg[2])
        elif msg[0] == '':
            with open('game_map.txt', 'a') as f:
                content = ''
                for item in coordinates:
                    content += item +','
                content += '\n'
                f.write(content) 
                coordinates = []
        else:
            # 
            with open('game_map.txt', 'r') as f:
                levels = f.readlines()
                # int() 
                i = int(msg[1]) - 1
                if i < len(levels):
                    send_message(levels[i].strip() + '')
                else:
                    send_message('')
```
