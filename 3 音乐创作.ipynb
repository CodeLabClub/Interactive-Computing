```python
# 创建列表 notes
notes = [1, 1, 5, 5, 6, 6, 5]
print(notes)
```

#### 获取列表元素
在音乐播放器这个项目中，我们要把简谱，如 1 1 5 5 6 6 5，逐个发送给 Scratch。我们可以用列表存放这些简谱，这些简谱称为列表的元素。获取列表元素时，我们要知道元素在列表中的位置（也称为索引）。



注意，Scratch 中列表的索引是从 1 开始，而 Python 是从 0 开始。Python 索引也可以从列表末尾开始，列表最后一个元素的索引是 -1，倒数第二个元素的索引是 -2，依此类推。

```python
# 获取列表元素
notes = [1, 1, 5, 5, 6, 6, 5]
print(notes[2])
print(notes[-2])
```

#### for 循环
这里我们要逐个发送歌曲列表中的简谱，可以用 for 循环来逐个获取列表中的元素，来认识一下吧！



```python
notes = [1, 1, 5, 5, 6, 6, 5]
for note in notes:
    print(note)
```

## 随机音乐
如果能够随机发送简谱，那么就可以生成随机音乐，这里我们需要使用生成随机数的函数以及能够重复发送生成音符的 for 循环。random.randint(a, b) 可以生成一个随机整数 N（a<=N<=b），for 循环如何让一段程序运行特定次数呢？

```python
# 随机音乐，可随机组合音乐片段
import time
import random
from codelab_adapter_client.message import send_message

for i in range(24):  # 重复执行 24 次
    note = random.randint(1, 7)  # 从 1-7 中随机生成一个整数 
    send_message(note)
    time.sleep(0.3)
```

### 编程百科
#### 循环特定次数
range(start, stop, step) 可以生成一个以 start 为起点，stop 为终点，步长为 step 的数字序列，它经常和 for 循环搭配使用来指定循环的次数。



```python
for i in range(10):
    print(i)
    print('hello')
```

```python
for i in range(0, 10, 2):
    print(i)
```

如果省略 step，则代表步长为 1，如果省略 start，则代表从 0 开始。

```python
for i in range(1, 3):
    print(i)
print('============')
for j in range(3):
    print(j)
```

## 创作音乐
音乐其实就是音符的组合，现在尝试用简谱写一段旋律，然后播放出来，看看听起来怎么样？你可以从编写一个音乐片段开始，或者改编你熟悉的歌曲，不要担心旋律是不是好听，先玩起来。

```python
# 音乐创作示例
import time
from codelab_adapter_client.message import send_message

p1 = [1, 3, 1, 3, 1, 3, 5, 0]
p2 = [2, 4, 2, 4, 2, 4 ,6, 0]
p3 = [1, 5, 3, 5, 1, 5, 3, 0]
p4 = [2, 6, 4, 6, 2, 6 ,4, 0]
# 列表元素可以是字符串、数字甚至列表，音乐中的旋律经常会重复
song3 = [p1, p2, p3, p4, p1, p2]  

# 用嵌套的 for 循环来获取列表中的元素
for p in song3:
    for note in p:
        send_message(note)
        time.sleep(0.3)
```
