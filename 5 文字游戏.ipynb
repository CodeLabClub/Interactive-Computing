### 复读机
先来体验一个复读机的文字游戏，打开项目[【交互计算】复读机](https://create.codelab.club/projects/8897/editor)，运行以下代码块。

```python
#【交互计算】复读机 https://create.codelab.club/projects/8897/editor
from codelab_adapter_client.message import send_message, receive_message

while True:
    # receive_message() 可以获取从 Scratch 发送过来的消息
    msg = receive_message()
    if msg:
        print(msg)
        msg = msg * 3 
        send_message(msg)
```

### 编程百科
#### while 循环
这里的程序要一直接收从 Scratch 发过来的消息，所以接收消息的代码要一直运行，这时候可以使用 while 循环。



for 循环是固定次数的循环，而 while 循环是条件循环，即根据条件是否满足来确定是否循环。

```python
a = 1
while a < 6:
    print(a)
    # 将 a 加 1 的结果赋值给 a，即 a 增加 1
    a = a + 1
```

while True 代表条件一直成立，所以循环会一直运行。如果想中断循环，可以使用 break。

```python
a = 1
while True:
    print(a)
    if a > 5:
        break
    a = a + 1
```

#### 编程解密
a = a + 1 这种形式可以简写成 a += 1 这种形式。类似的还有：

| 运算 | 等价形式 | 
|------|------|
|  a = a + 2  | a += 2 |
|  a = a - 2  | a -= 2 |
|  a = a * 2  | a *= 2 |
|  a = a / 2  | a /= 2 |

#### 字符串运算
字符串进行乘法运算时，相当于将字符串进行复制：

```python
print(6)
print('=' * 20)  # 打印分隔线
print(666)
```

## 记单词
接下来我们一起来制作一个记单词的项目，它会显示一个词语的中英文和图片，并且朗读单词。

```python
#【交互计算】记单词 https://create.codelab.club/projects/9186/editor/
from codelab_adapter_client.message import send_message, receive_message

# 用字典来存放单词的中英文
words = {'苹果':'apple', '香蕉':'banana', '面包':'bread', '西瓜':'watermelon', '牛奶':'milk', '蛋糕':'cake', '鸡蛋':'egg'}
while True:
    msg = receive_message()
    if msg:
        send_message(msg + ' ' + words[msg])
```

对这个项目进行升级，实现单词测验的功能。

```python
#【交互计算】记单词升级 https://create.codelab.club/projects/9912/editor
from codelab_adapter_client.message import send_message, receive_message

words = {'苹果':'apple', '香蕉':'banana', '面包':'bread', '西瓜':'watermelon', '牛奶':'milk', '蛋糕':'cake', '鸡蛋':'egg'}
while True:
    msg = receive_message()
    if msg:
        send_message(words[msg])
```

### 编程百科
#### 字典
这里使用了字典 (dictionary) 这种数据结构来存放中文和英文单词，字典适合用在根据一个值查找另一个值的情况，就像我们使用中英文字典查找一个中文词语对应的英文单词。



```python
words = {'苹果':'apple', '香蕉':'banana', '面包':'bread'}
# 根据键查找对应的值
print(words['苹果'])  
# 把数据放入字典
words['芒果'] = 'mango'
print(words)
```
