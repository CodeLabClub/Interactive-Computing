## JupyterLab 快捷键
使用快捷键可以更方便地操作单元格，使用快捷键时，先点击选中单元格，以下为 JupyterLab 中部分常用的快捷键：
- A：在选中单元格的上方插入新单元格
- B：在选中单元格的下方插入新单元格
- D，D：按两次 D 键删除单元格
- Shift + Enter：运行单元格
- M：进入文本编辑模式
- Y：进入代码编辑模式
